```python
import matplotlib.pyplot as plt
import random
from time import time
import numpy as np
```

```python
def Burbuja(a,n):
    for i in range(1,n):
        for j in range(0,n-i):
            if(a[j] > a[j+1]):
                k = a[j+1]
                a[j+1] = a[j]
                a[j] = k;
```

```python
def Insercion(a,n):
    for i in range(1,n):
        v=a[i]
        j=i-1
        while j >= 0 and a[j] > v:
            a[j+1] = a[j]
            j=j-1
        a[j+1]=v
```

```python
def Seleccion(a,n):
    for i in range(0,n-1):
        min=i
        for j in range(i+1,n):
            if a[min] > a[j]:
                min=j
        aux=a[min]
        a[min]=a[i]
        a[i]=aux
```

```python
def GeneraR(n):
    a=[]
    for i in range(0,n):
        a.append(random.randrange(0, n, 1))
    return a
```

```python
def GeneraD(n):
    a=[]
    for i in range(0,n):
        a.append(i)
    return a  
```

```python
def GeneraI(n):
    a=[]
    for i in range(0,n):
        a.append(n-i)
    return a   
```

```python
def imprime(a,n):
    for i in range(0,n):
        print (a[i])
```

En cada iteración, generamos un vector directo `[1,2,3,4...]`, un vector inverso ``[..., 4,3,2,1]``,y un vector que genera números aleatorios.

```python
def experimento(n):
    
        V1 =[]  # ALEATORIO
    # Orden directo
        V2=GeneraD(n)
    # Orden inverso
        V3=GeneraI(n)
    # Experimento aleatorio
        tb = 0 # BURBUJA
        ti = 0 # INSERCION
        ts = 0 # SELECCION


        for _ in range(10):
# Aleatorio. 10 experimentos
# Realizar experimentos y completar las listas usando GeneraR(n)
            V1 = GeneraR(n)
            # BURBURJA
            t1 = time()
            Burbuja(V1,n)
            t2 = time() - t1
            tb += t2
            # INSERCION
            t1 = time()
            Insercion(V1,n)
            t2 = time() - t1
            ti += t2
            # SELECCION
            t1 = time()
            Seleccion(V1,n)
            t2 = time() - t1
            ts += t2
        TB1.append(int(1000*tb/10))
        TI1.append(int(1000*ti/10))
        TS1.append(int(1000*ts/10))

        
            

# Experimento directo
        # BURBURJA
        tb = time()
        Burbuja(V2,n)
        tb = time() - tb
        TB2.append(int(1000*tb))
        # INSERCION
        ti = time()
        Insercion(V2,n)
        ti = time() - ti
        TI2.append(int(1000*ti))
        # SELECCION
        ts = time()
        Seleccion(V2,n)
        ts = time() - ts
        TS2.append(int(1000*ts))

# Experimento inverso
        # BURBURJA
        tb = time()
        Burbuja(V3,n)
        tb = time() - tb
        TB3.append(int(1000*tb))
        # INSERCION
        ti = time()
        Insercion(V3,n)
        ti = time() - ti
        TI3.append(int(1000*ti))
        # SELECCION
        ts = time()
        Seleccion(V3,n)
        ts = time() - ts
        TS3.append(int(1000*ts))
```

El vector Tam tiene los números del 1 al 20, y los tamaños de los arrays que tenemos que probar van del 100 al 2000, añadiendo 100 elementos en cada iteracción, por lo que, en cada iteracción, el tamaño de los arrays es el del correspondiente elemento del vector Tam multiplicado por 100.
```python
def QuickSort(a,iz,de, PIVOTE_A_LA_MEDIANA = True):
    i=iz
    j=de
    #
    # Selección del pivote
    # Elegir uno a descomentar
    #
    # Pivote en la mediana
    if PIVOTE_A_LA_MEDIANA:
        x=a[int((iz + de)/2)]
    # Pivote en el lado izquierdo
    if not PIVOTE_A_LA_MEDIANA:
        x=a[iz]
 
    while( i <= j ):
        while a[i]<x and j<=de:
            i=i+1
        while x<a[j] and j>iz:
            j=j-1
        if i<=j:
            aux = a[i]; a[i] = a[j]; a[j] = aux;
            i=i+1;  j=j-1;
 
    if iz < j:
        QuickSort( a, iz, j ,PIVOTE_A_LA_MEDIANA);
    if i < de:
        QuickSort( a, i, de ,PIVOTE_A_LA_MEDIANA);
```

```python
def MergeSort(a,n):
    if n>1:
        m = n//2;
        l = a[:m]
        r = a[m:]

        MergeSort(l,len(l))
        MergeSort(r,len(r))

        i=0; j=0; k=0
        while i < len(l) and j < len(r):
            if l[i] < r[j]:
                a[k]=l[i]
                i=i+1
            else:
                a[k]=r[j]
                j=j+1
            k=k+1

        while i < len(l):
            a[k]=l[i]
            i=i+1
            k=k+1

        while j < len(r):
            a[k]=r[j]
            j=j+1
            k=k+1
```
